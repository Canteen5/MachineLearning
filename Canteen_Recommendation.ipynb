```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
menu = '/content/drive/MyDrive/fake_menu.csv'
order = '/content/drive/MyDrive/fake_orders.csv'

menu_df = pd.read_csv(menu)
order_df = pd.read_csv(order)
```

```python
#menu_df.head()
order_df.head()
```

```python
global_counts = (
    order_df.groupby('item_id')['quantity'].sum().rename('global_count').reset_index()
)# This counts the frequency of each food item ordered. (item_id 5 was ordered 100 times by --- customers in total)
```

```python
global_counts.head() # global_count displays the quantity of each item_id in total.
```

```python
global_counts['global_count'] = global_counts['global_count']/global_counts['global_count'].sum() #global_count will store the division of the individual quantities (global_count) to the sum of all the global_count.
```

```python
user_count = (
    order_df.groupby(['user_id','item_id'])['quantity']
    .sum()
    .reset_index()
)
user_total = (
    order_df.groupby('user_id')['quantity']
    .sum()
    .rename('user_t')
    .reset_index()
)
user_count_score = user_count.merge(user_total, on='user_id')
user_count_score['user_score'] = user_count_score['quantity']/user_count_score['user_t']
```

```python
user_count_score.head(), global_counts.head()
```

```python
menu_df.head()
```

```python
global_counts.merge(menu_df, on='item_id')
```

```python
global_maps = global_counts.set_index('item_id')['global_count'].to_dict()
```

```python
user_map = {}
for _, row in user_count_score.iterrows():
  user_map[(int(row['user_id']), int(row['item_id']))] = row['user_score']
```

```python
print(type(user_map))
print(len(user_map))
list(user_map.keys())[:10]
```

```python
list(user_map.keys())[:5]
```

```python
def recommend (user_id, k=5, alpha = 0.6, beta = 0.4):
 rows = []
 for _, item in menu_df.iterrows():
  iid = item['item_id']
  g = global_maps.get(iid, 0)
  u = user_map.get((user_id, iid), 0)
  score = alpha*g + beta*u
  rows.append({
    "item_id": iid,
    "item_name": item["item_name"],
    "category": item["category"],
    "price": item["price"],
    "global_score": g,
    "user_score": u,
    "hybrid_score": score,
})
 df = pd.DataFrame(rows)
 return df.sort_values("hybrid_score", ascending=False).head(k)
```

```python
print(type(global_maps))
print(type(user_map))
```

```python
print(recommend(10))
```

```python
while True:
  try:
    uid = int(input("Enter the user_id to get Recommendation!"))
    if uid == -1:
      break
    print(recommend(user_id = uid))
  except ValueError:
    print("Enter a number!")
```
